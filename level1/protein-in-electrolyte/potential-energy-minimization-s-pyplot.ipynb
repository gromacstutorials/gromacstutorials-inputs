```python
import numpy as np
import sys, os, git
from matplotlib import pyplot as plt
```

```python
current_path = os.getcwd()
git_repo = git.Repo(current_path, search_parent_directories=True)
git_path = git_repo.git.rev_parse("--show-toplevel")
sys.path.append(git_path + "/pyplot-perso/")
from functions import complete_panel, save_figure, set_boundaries, \
                      add_subplotlabels, set_boundaries, prepare_figure
from color_series1 import colors
path_figures = current_path[len(git_path):] + '/'
```

```python
t, energy = np.loadtxt("potential-energy-minimization-s.xvg", skiprows=24).T
```

```python
filename = "potential-energy-minimization-s"
for mode, mygray in zip(['light', 'dark'], [colors["mylightgray"], colors["mydarkgray"]]):    
    fig = prepare_figure(mode, transparency = True, desired_figsize=(18,6))
    ax, n, l_tot, c_tot = [], 0, 1, 1
    n += 1
    ax.append(plt.subplot(l_tot, c_tot, n))
    ax[-1].plot(t, energy/1000, "o", color=colors["mycyan"],
                markersize = 12, linewidth=4)
    complete_panel(ax[-1], r'$N_\mathrm{step}$', r'$E_\mathrm{p} ~ \mathrm{(MJ/mol)}$', xpad=10,
                   legend=False, axis_color=mygray)
    set_boundaries(plt, x_ticks=np.arange(0, 52, 10), x_boundaries=(-2, 52), 
                   y_boundaries=(-225, 0), y_ticks=np.arange(-200, 1, 50))
    #add_subplotlabels(fig, ax, ["a", "b"], color=mygray)
    save_figure(plt, fig, mode, git_path, path_figures, filename)
```
