```python
import numpy as np
import sys, os, git
from matplotlib import pyplot as plt
```

```python
current_path = os.getcwd()
git_repo = git.Repo(current_path, search_parent_directories=True)
git_path = git_repo.git.rev_parse("--show-toplevel")
sys.path.append(git_path + "/pyplot-perso/")
from functions import complete_panel, save_figure, set_boundaries, \
                      add_subplotlabels, set_boundaries, prepare_figure
from color_series1 import colors
path_figures = current_path[len(git_path):] + '/'
```

```python
filename = "rdf-production"
for mode, mygray in zip(['light', 'dark'], [colors["mylightgray"], colors["mydarkgray"]]):    
    fig = prepare_figure(mode, transparency = True, desired_figsize=(18,18))
    ax, n, l_tot, c_tot = [], 0, 3, 1
    n += 1
    ax.append(plt.subplot(l_tot, c_tot, n))
    x, y = np.loadtxt("sol-sol-rdf.xvg", skiprows=25).T
    ax[-1].plot(x*10, y, "o", color=mygray,
                markersize = 12, linewidth=4, label=r"$\textrm{H}_2\textrm{O}-\textrm{H}_2\textrm{O}$")
    x, y = np.loadtxt("na-sol-rdf.xvg", skiprows=25).T
    ax[-1].plot(x*10, y, "o", color=colors["mycyan"],
                markersize = 12, linewidth=4, label=r"$\textrm{Na}^+-\textrm{H}_2\textrm{O}$")
    x, y = np.loadtxt("so4-sol-rdf.xvg", skiprows=25).T
    ax[-1].plot(x*10, y, "o", color=colors["myorange"],
                markersize = 12, linewidth=4, label=r"$\textrm{SO}_4^{2-}-\textrm{H}_2\textrm{O}$")
    complete_panel(ax[-1], r'$r (\textrm{\AA}$)', r'$\mathrm{R.D.F}$',
                   xpad=10, legend=True, axis_color=mygray)
    set_boundaries(plt, x_boundaries=(0.8, 4.2), y_boundaries=(0, 3.5))
                   # y_ticks=np.arange(-1.00, 3.501, 1.000)) x_ticks=np.arange(0, 22, 5), 
    #add_subplotlabels(fig, ax, [r"$\textrm{a}$", r"$\textrm{b}$", r"$\textrm{c}$"], color=mygray)
    save_figure(plt, fig, mode, git_path, path_figures, filename)
```
