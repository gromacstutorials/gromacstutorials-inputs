```python
import numpy as np
import sys, os, git
from matplotlib import pyplot as plt
```

```python
current_path = os.getcwd()
git_repo = git.Repo(current_path, search_parent_directories=True)
git_path = git_repo.git.rev_parse("--show-toplevel")
sys.path.append(git_path + "/pyplot-perso/")
from functions import complete_panel, save_figure, set_boundaries, \
                      add_subplotlabels, set_boundaries, prepare_figure
from color_series1 import colors
path_figures = current_path[len(git_path):] + '/'
```

```python
filename = "msd-production"
for mode, mygray in zip(['light', 'dark'], [colors["mylightgray"], colors["mydarkgray"]]):    
    fig = prepare_figure(mode, transparency = True, desired_figsize=(18,22))
    ax, n, l_tot, c_tot = [], 0, 3, 1
    n += 1
    ax.append(plt.subplot(l_tot, c_tot, n))
    x, y = np.loadtxt("sol-msd.xvg", skiprows=25).T
    ax[-1].loglog(x*10, y, "o", color=mygray,
                markersize = 12, linewidth=4, label=r"$\textrm{H}_2\textrm{O}$")
    x, y = np.loadtxt("so4-msd.xvg", skiprows=25).T
    ax[-1].loglog(x*10, y, "o", color=colors["myorange"],
                markersize = 12, linewidth=4, label=r"$\textrm{SO}_4^{2-}$")
    x, y = np.loadtxt("na-msd.xvg", skiprows=25).T
    ax[-1].loglog(x*10, y, "o", color=colors["mycyan"],
                markersize = 12, linewidth=4, label=r"$\textrm{Na}^+$")
    x = np.logspace(2,3)
    ax[-1].loglog(x, x/150, "--", color=mygray,
                markersize = 12, linewidth=4, label=r"$\propto t$")

    complete_panel(ax[-1], r'$t (\textrm{ps})$', r'$\textrm{msd} ~ (\textrm{nm}^2)$',
                   xpad=15, legend=True, axis_color=mygray)

    set_boundaries(plt, x_boundaries=(6, 6000), y_boundaries=(0.008, 20))
                   # y_ticks=np.arange(-1.00, 3.501, 1.000)) x_ticks=np.arange(0, 22, 5), 
    #add_subplotlabels(fig, ax, [r"$\textrm{a}$", r"$\textrm{b}$", r"$\textrm{c}$"], color=mygray)
    save_figure(plt, fig, mode, git_path, path_figures, filename)
```
